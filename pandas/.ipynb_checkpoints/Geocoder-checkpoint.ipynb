```python
from geopy.geocoders import Nominatim
dir(Nominatim)
```

```python
import pandas
df = pandas.read_csv("Supermarkets.csv")
df
```

```python
nom = Nominatim()
df["Address"] = df["Address"] + ", " + df["City"] + ", " + df["State"] + ", " + df["Country"]
df
```

```python
# How to apply a method iteratively on all columns together
```

```python
df["Coordinates"] = df["Address"].apply(nom.geocode)  #applys nom.geocode("Adress string") on all values of address 
```

```python
df # Yay i got coordinates
```

```python
df["Coordinates"].apply(lambda x: x.latitude)
```

```python
# now trying to get latitude :
df["Coordinates"].apply(lambda x : x.latitude if x != None else None)  # Lambdas are inline functions
```

```python
df["Latitude"] = df["Coordinates"].apply(lambda x : x.latitude if x != None else None)
df
```

```python
df["Longitude"] = df["Coordinates"].apply(lambda x : x.longitude if x != None else None)
df
```
